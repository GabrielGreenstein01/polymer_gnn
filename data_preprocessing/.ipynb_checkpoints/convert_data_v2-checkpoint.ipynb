```python
import pandas as pd
import numpy as np
import re
```

```python
df_grampa_unfiltered = pd.read_csv("./labeled_data/grampa_positives.csv")
df_grampa_unfiltered.shape
```

```python
# Drop duplicates in unfiltered
df_grampa_unfiltered = df_grampa_unfiltered.drop_duplicates(subset=['sequence'])
df_grampa_unfiltered = df_grampa_unfiltered.reset_index()
df_grampa_unfiltered.shape
```

```python
df_grampa_filtered = pd.read_csv("./labeled_data/July10th_FilteredGRAMPA.csv")
df_grampa_filtered.head()
```

```python
# Find rows in unfiltered df whose sequences are in filtered df
sequences = df_grampa_filtered['sequence'].to_list()
df_grampa_unfiltered = df_grampa_unfiltered[df_grampa_unfiltered['sequence'].isin(sequences)]
df_grampa_unfiltered.to_csv(r'test.csv', index=None, sep=',', mode='w')
```
