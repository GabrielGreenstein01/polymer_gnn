```python
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors, rdDepictor
from IPython.display import display, Image
```

```python
def seq_to_SMILES(seq):
    mol = Chem.MolFromFASTA(seq)
    smiles = Chem.MolToSmiles(mol, canonical=True, isomericSmiles=False)
    print(smiles)
    return display(Draw.MolToImage(mol, size=(500,500)))
```

```python
seq_to_SMILES('AAWGRMSTELP')
```

```python
seq_to_SMILES('VLDILHFYTKPA')
```

```python
seq_to_SMILES('EFYNAAQGCW')
```
