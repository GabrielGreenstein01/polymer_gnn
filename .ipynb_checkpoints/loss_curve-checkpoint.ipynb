```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
LOG_PATH = 'past_trials/AttentiveFP/100_epochs'
df = pd.read_csv(LOG_PATH + '/train_log.txt')
```

```python
epochs = df['epoch'].unique()

train_loss = df[df['type'] == 'train']['loss'].values
val_loss = df[df['type'] == 'val']['loss'].values
inference_loss = df[df['type'] == 'inference']['loss'].values
```

```python
# Create a figure and two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Plot for train loss
ax[0].plot(epochs, train_loss, marker='o', linestyle='-', color='r', label='train', markersize=3)
ax[0].plot(epochs, val_loss, marker='o', linestyle='-', color='g', label='val', markersize=3)
ax[0].plot(epochs, inference_loss, marker='o', linestyle='-', color='b', label='inference', markersize=3)
ax[0].set_title('w/ Inference Loss')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].legend()  # Add legend to the first subplot

# Plot for validation loss
ax[1].plot(epochs, train_loss, marker='o', linestyle='-', color='r', label='train', markersize=3)
ax[1].plot(epochs, val_loss, marker='o', linestyle='-', color='g', label='val', markersize=3)
ax[1].set_title('w/o Inference Loss')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Loss')
ax[1].legend()  # Add legend to the second subplot

# Adjust the layout
plt.tight_layout()

plt.savefig(LOG_PATH + '/loss_fig.png')


# Show the plots
# plt.show()
```

```python
LOG_PATH = 'past_trials/AttentiveFP/100_epochs'
df = pd.read_csv(LOG_PATH + '/train_log.txt')
train = df.loc[(df['type'] == 'train') & (df['loss'].isnull() == False)].drop(['type', 'elapsed_time'], axis=1)
val = df.loc[(df['type'] == 'val') & (df['loss'].isnull() == False)].drop(['type', 'elapsed_time'], axis=1)
# infer = df.loc[(df['type'] == 'inference') & (df['loss'].isnull() == False)].drop(['type', 'elapsed_time'], axis=1)
```

```python
train_sum = train.groupby('epoch')['loss'].sum().reset_index()
val_sum = val.groupby('epoch')['loss'].sum().reset_index()
# infer_sum = infer.groupby('epoch')['loss'].sum().reset_index()
```

```python
train_sum = train_sum['loss'].apply(lambda x: x/max(train['batch'].unique()))
val_sum = val_sum['loss'].apply(lambda x: x/max(val['batch'].unique()))
# infer_sum = infer_sum['loss'].apply(lambda x: x/max(infer['batch'].unique()))
```

```python
x = df['epoch'].unique()

plt.plot(x, train_sum, color='r', label='train') 
plt.plot(x, val_sum, color='g', label='val')
# plt.plot(x, infer_sum, color='b', label='inference') 
  
# Naming the x-axis, y-axis and the whole graph 
plt.xlabel("batch / epoch") 
plt.ylabel("loss") 
# plt.title("Train vs. Loss") 
  
# Adding legend, which helps us recognize the curve according to it's color 
plt.legend() 

plt.savefig(LOG_PATH + '/loss_fig.png')
# To load the display window 
plt.show() 
```
